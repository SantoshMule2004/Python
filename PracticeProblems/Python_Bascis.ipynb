### - Control Flow

```python
# Conditional statements
if a > b:
    print(f"{a} is > {b}")
elif a < b:
    print(f"{a} is < {b}")
else:
    print(f"{a} is equal {b}")
```

```python
# ternary 
res = "a is > b" if a > b else "a is < b"
print(res)
```

```python
# Looping statements
# 1. for loops

for num in nums:
    print(num)

for i in range(len(nums)):
    print(nums[i])

for index, num in enumerate(nums):
    print(f"{index} : {num}")
    if index == 2:
        break
else:
    print("Executed successfully")
```

```python
# While loop
while b < a:
    print(b)
    b += 1
```

```python
# Jump statements
for num in nums:
    if num == 1:
        continue
    
    print(num)

    if num == 4:
        break
```

```python
# Switch-case
def weekDay(num: int):
    switch = {
        1: "Sunday",
        2: "Monday",
        3: "Tuesday",
        4: "Wednesday",
        5: "Thurday",
        6: "Friday",
        7: "Saturday"
    }
    return switch.get(num, "Invalid Day")

weekDay(7)
```

```python
# Match-case
num = 1
match num:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case _:
        print("Invalid day")
```

## Functions

```python
# function
def add(a, b):
    return a + b
print(add(1, 2))
```

```python
# Closures
def outerFunc(num):
    count = 1
    def innerFunc():
        return count + num
    return innerFunc

func = outerFunc(12)
print(func.__closure__[1].cell_contents)
```

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x += 1
        return x
    return inner

inn = outer()
print(inn())
print(inn())
print(inn())
print(inn())
print(inn())
print(inn())
print(inn())
print(inn())
```

```python
# Decorators
def debug(original_function):
    def wrapper_function():
        print("Before the original function runs")
        original_function()
        print("After the original function runs")
    return wrapper_function

@debug
def display():
    print("This is the original function")

display()
```

```python
def log(func):
    def wrapper(*args, **kwargs):
        print(f"Function '{func.__name__}' called with args={args} and kwargs={kwargs}")
        return func(*args, **kwargs)
    return wrapper

@log
def addNum(a, b):
    return a + b

print(addNum(2, 3))
```

```python
# docstings
def multiply(a, b):
    """This function returns the multiplication of two numbers"""
    return a * b

print(multiply.__doc__)
print(multiply(2, 3))
```

## Object-Oriented Programming

```python
class Info:
    def __init__(self, name="Alice", age=19):
        self._name = name # protected
        self.__age = age # private

    def __repr__(self):
        return f"name={self._name}, age={self.__age}"

student = Info("Santosh", 21)
print(student)
print(student._name)
print(student._Info__age) 
```

```python
# Abstraction
from abc import ABC, abstractmethod

class car(ABC):
    def __init__(self, carname):
        super().__init__()
        self.carName = carname
    
    def drive(self):
        print(f"Driving {self.carName}")
    
    @abstractmethod
    def horn(self):
        pass

class Swift(car):
    def __init__(self, carname):
        super().__init__(carname)

    def horn(self):
        print(f"POm pom")

s = Swift("Swift")
s.drive()
s.horn()
```

```python
# multiple inheritance
class A():
    def __init__(self, x, **kwargs):
        print("A init:", x)
        super().__init__(**kwargs)

    def show(self):
        print(f"Showing from A {self.x}")

class B():
    def __init__(self, y, **kwargs):
        print("B init:", y)
        super().__init__(**kwargs)

    def show(self):
        print(f"Showing from B {self.y}")

class C(B, A):
    def __init__(self, y, **kwargs):
        super().__init__(y, **kwargs)

c = C(y=3, x=4)
```

```python
# polymorphism
def add(a, b):
    return a + b # operator overloading

print(add(10, 20))     # int + int
print(add("Hello", " World"))  # str + str
print(add([1, 2], [3, 4]))    # list + list
```

```python
# Duck typing
class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

def animal_sound(animal):
    print(animal.speak())

animal_sound(Dog())
animal_sound(Cat())
```

```python
# Method overriding
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

p = Penguin()
p.fly()
```

```python
# operator overloading
class Point:
    def __init__(self, x, y):
        self.x, self.y = x, y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(2, 3)
p2 = Point(4, 5)
print(p1 + p2)
```

```python
class dummy():
    def __init__(self):
        pass

    @staticmethod
    def show():
        print("This is static method")


class D(dummy):
    def __init__(self):
        super().__init__()

    def show(show):
        print("cannot ovveride static metod")

d = D()
print(d.show())
```

## Iterators and generators

```python
itr = iter(nums)
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
print(next(itr))
```

```python
# Generators
def count_up_to(n):
    num = 1
    while num <= n:
        yield num
        num += 1

for val in count_up_to(5):
    print(val)
```

```python
gen = (x * x for x in range(5))
# print(next(gen))  # 0
# print(next(gen))  # 1
print(list(gen))  # [4, 9, 16]
```

```python
nums = [1, 2, 3]
it = iter(nums)

while True:
    try:
        val = next(it)
        print(val)
    except StopIteration:
        print("End of iteration reached")
        break
```

```python
def gen():
    yield 1
    yield 2

g = gen()
print(next(g))  # 1
print(next(g))  # 2
# print(next(g))  # Raises StopIteration
```

## Exception handling

```python
try:
    # Code that might cause an error
    x = 10 / 0
except ZeroDivisionError:
    # Code that runs if an exception occurs
    print("Cannot divide by zero")
else:
    # Runs if no exception occurs
    print("Division successful")
finally:
    # Runs no matter what
    print("Execution complete")
```
